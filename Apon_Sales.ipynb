```python
df2.rename(columns={"Total Sales Amount": "Month_1_Customer_Sales"}, inplace=True)
df3.rename(columns={"Total Sales Amount": "Month_2_Customer_Sales"}, inplace=True)
df4.rename(columns={"Total Sales Amount": "Month_3_Customer_Sales"}, inplace=True)
```

```python
df1.rename(columns={"Apon ID": "Customer Code"}, inplace=True)
```

```python
df = pd.merge(df3,df2, how='outer', on = 'Customer Code', suffixes = ('_left', '_right'))
df
```

```python
df = pd.merge(df,df4, how='outer', on = 'Customer Code', suffixes = ('_left', '_right'))
df
```

```python
df = pd.merge(df,df1, how='outer', on = 'Customer Code', suffixes = ('_left', '_right'))
df
```

```python
df.columns
```

```python
df.drop(['SL_left', 'SL_right', 'SL', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6',
       'Unnamed: 7' ], axis =1, inplace = True)
```

```python
df['Month_1_Customer_Sales'] = df['Month_1_Customer_Sales'].fillna(0)
df['Month_2_Customer_Sales'] = df['Month_2_Customer_Sales'].fillna(0)
df['Month_3_Customer_Sales'] = df['Month_3_Customer_Sales'].fillna(0)
```

```python
df.head()
```

```python
df['Total sales'] = df['Month_1_Customer_Sales'] + df['Month_2_Customer_Sales'] + df['Month_3_Customer_Sales']
```

```python
df.head()
```

```python
df.isnull().sum() * 100 / len(df)
```

```python
df = df[df['Total sales'] != 0.0]
```

```python
len(df)
```

```python
df.head()
```

```python
df.insert(6,"Month 1 Customer", " ")
df['Month 1 Customer'] = np.where(df['Month_1_Customer_Sales'] == 0.0, 'no', 'yes')
```

```python
df.insert(7,"Month 2 Customer", " ")
df['Month 2 Customer'] = np.where(df['Month_2_Customer_Sales'] == 0.0, 'no', 'yes')
```

```python
df.insert(8,"Month 3 Customer", " ")
df['Month 3 Customer'] = np.where(df['Month_3_Customer_Sales'] == 0.0, 'no', 'yes')
```

```python
df.insert(9,"Regularity", " ")
```

```python
df["Regularity"] = np.where( ((df['Month 1 Customer'] == 'yes') & (df['Month 2 Customer'] == 'yes') & (df['Month 3 Customer'] == 'yes')),'3 Months',
(np.where((((df['Month 1 Customer'] == 'yes') & (df['Month 2 Customer'] == 'yes'))| ((df['Month 2 Customer'] == 'yes') & (df['Month 3 Customer'] == 'yes')) | ((df['Month 3 Customer'] == 'yes') & (df['Month 1 Customer'] == 'yes'))), '2 Months', '1 Month')))

#(((df['Month 1 Customer'] == 'yes') & (df['Month 2 Customer'] == 'yes'))| ((df['Month 2 Customer'] == 'yes') & (df['Month 3 Customer'] == 'yes')) | ((df['Month 3 Customer'] == 'yes') & (df['Month 1 Customer'] == 'yes')))
```
